```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the prices data
df = pd.read_csv('prices_round_1_day_0.csv', sep=';')

# Get the unique products available in the dataset
products = df['product'].unique()

# Set up the plotting area
plt.figure(figsize=(14, 8))

# Create a separate subplot for each product
for i, product in enumerate(products):
    product_df = df[df['product'] == product]
    
    plt.subplot(len(products), 1, i + 1)
    
    # Scatter plot for ask prices (in blue)
    plt.scatter(product_df['timestamp'], product_df['ask_price_1'], color='blue', label='Ask Price 1', s=5, alpha=0.5)
    
    # Scatter plot for bid prices (in red)
    plt.scatter(product_df['timestamp'], product_df['bid_price_1'], color='red', label='Bid Price 1', s=5, alpha=0.5)
    
    plt.title(f'Bid and Ask Prices for {product}')
    plt.xlabel('Timestamp')
    plt.ylabel('Price')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the prices data
df = pd.read_csv('prices_round_1_day_0.csv', sep=';')

# Get the unique products available in the dataset
products = df['product'].unique()

# Loop through each product and create a separate graph
for product in products:
    product_df = df[df['product'] == product]
    
    # Initialize a new figure for the product
    fig = go.Figure()
    
    # Add a scatter trace for Ask Prices in blue
    fig.add_trace(go.Scatter(
        x=product_df['timestamp'],
        y=product_df['ask_price_1'],
        mode='markers',
        marker=dict(color='blue', size=4, opacity=0.5),
        name='Ask Price 1'
    ))
    
    # Add a scatter trace for Bid Prices in red
    fig.add_trace(go.Scatter(
        x=product_df['timestamp'],
        y=product_df['bid_price_1'],
        mode='markers',
        marker=dict(color='red', size=4, opacity=0.5),
        name='Bid Price 1'
    ))
    
    # Update layout with titles
    fig.update_layout(
        title=f'Bid and Ask Prices for {product}',
        xaxis_title='Timestamp',
        yaxis_title='Price',
        template='plotly_white'
    )
    
    # This command renders the plot as an HTML file and opens it in your default browser
    fig.show()
```
